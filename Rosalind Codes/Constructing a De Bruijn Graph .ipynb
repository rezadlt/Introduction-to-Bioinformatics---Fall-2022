```python
file = open("/Users/mac/Desktop/courses/Bioinformatic/a.txt", "r")
set_S = set()
for line in file:
    set_S.add(line.strip())
file.close()
    
def Node(sequence):
    complement = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    b_k = ''
    for j in sequence[::-1]:
        b_k += complement[j]
    return b_k

def main():
    set_S_reverse_complement = set()
    for i in set_S:
        set_S_reverse_complement.add(Node(i))
    set_U = set_S | set_S_reverse_complement
    edges = set()
    for i in set_U:
        edges.add((i[:len(i) - 1], i[1:]))
    for i in edges:
        output = '(' + i[0] + ", " + i[1] + ')'
        print(output)

if __name__ == "__main__":
    main()
```
