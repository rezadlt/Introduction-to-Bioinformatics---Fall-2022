```python
def rosalind_corr(path):
    with open(path) as fp:
        sequences=[f.strip() for f in fp if f[0]!='>']
    return error_correction(sequences)

def reverse_complement(dna):
    translate={'A':'T','T':'A','C':'G','G':'C'}
    return ''.join(translate[k] for k in dna)[::-1]

def hamming_distance(x,y):
    count = 0
    for i in range(len(x)):
        if x[i] != y[i]:
            count = count + 1
    return count

def error_correction(sequences):
    ref = [k for x in sequences for k in [x,reverse_complement(x)]]
    old_read = [k for k in sequences if ref.count(k)<2]
    new_read = [k for k in sequences if ref.count(k)>=2]
    res = []
    for x in old_read:
        for y in new_read:
            if hamming_distance(x,y) == 1:
                res.append(f'{x}->{y}')
                break
            if hamming_distance(x,reverse_complement(y)) == 1:
                res.append(f'{x}->{reverse_complement(y)}')
                break
    return '\n'.join(res)
print(rosalind_corr(r'/Users/mac/Desktop/courses/Bioinformatic/rosalind_corr.txt'))
```
