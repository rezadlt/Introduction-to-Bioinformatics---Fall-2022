```python
def char_to_num(ch):
    if ch == 'A':
        return 0
    elif ch == 'C':
        return 1
    elif ch == 'G':
        return 2
    else:
        return 3

def arg_min(arr):
    min_ = 100000000
    inedx = -1
    for i in range(len(arr)):
        if arr[i] < min_:
            min_ = arr[i]
            index = i
    return index

def num_to_char(number):
    ar = ['A', 'C', 'G', 'T']
    return ar[number]

score = [[0, 1, 1, 1], [1, 0, 1, 1], [1, 1, 0, 1], [1, 1, 1, 0]]
num = int(input())
strs = []
adj = []
for i in range(2 * num - 1):
    adj.append([])
for i in range(num):
    li = input().split("->")
    strs.append(li[1])
    ind1 = len(strs) - 1
    ind2 = int(li[0])
    adj[ind2].append(ind1)
for i in range(num - 2):
    li = list(map(int, input().split("->")))
    adj[li[0]].append(li[1])
for i in range(num - 1):
    strs.append("")
root = 2 * num - 2

def DFS(arr, v, index):
    if len(adj[v]) == 0:
        x = char_to_num(strs[v][index])
        for i in range(4):
            if i == x:
                arr[v][i] = 0
            else:
                arr[v][i] = 1000000000
        return
    for i in adj[v]:
        DFS(arr, i, index)
    for i in range(4):
        one, two = adj[v]
        arr[v][i] = min((arr[one][j] + score[i][j]) for j in range(4)) + \
                    min((arr[two][k] + score[i][k]) for k in range(4))

def char_compute(arr, v, ch):
    if len(adj[v]) == 0:
        return
    strs[v] += num_to_char(ch)
    one = arg_min([(arr[adj[v][0]][j] + score[ch][j]) for j in range(4)])
    two = arg_min([(arr[adj[v][1]][j] + score[ch][j]) for j in range(4)])
    char_compute(arr, adj[v][0], one)
    char_compute(arr, adj[v][1], two)

def smallParsinomy():
    global root
    arr = []
    for i in range(len(adj)):
        arr.append([0] * 4)
    for i in range(len(strs[0])):
        DFS(arr, root, i)
        nu = arg_min(arr[root])
        char_compute(arr, root, nu)

def diff(i, j):
    s = 0
    for k in range(len(strs[0])):
        if strs[i][k] != strs[j][k]:
            s += 1
    return s

smallParsinomy()
diffs = []
su = 0
for i in range(len(adj)):
    for t in adj[i]:
        dif = diff(i, t)
        su += dif
        diffs.append(dif)
nu = 0
obj = open("rosalind_ba7f_Out.txt", "w")
obj.write(str(su) + "\n")
for i in range(len(adj)):
    for t in adj[i]:
        obj.write(strs[i] + "->" + strs[t] + ":" + str(diffs[nu]) + "\n")
        obj.write(strs[t] + "->" + strs[i] + ":" + str(diffs[nu]) + "\n")
        nu += 1
obj.close()
```
