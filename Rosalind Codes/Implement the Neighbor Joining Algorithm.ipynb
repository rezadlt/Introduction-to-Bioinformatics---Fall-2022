```python
nu = 0

def compute_Q(score):
    tdis = []
    for i in range(len(score)):
        tdis.append(sum(score[i][j] for j in range(len(score))))
    ret = []
    for i in range(len(score)):
        ret.append([])
        for j in range(len(score)):
            if j < i:
                ret[i].append(ret[j][i])
            elif j == i:
                ret[i].append(0)
            else:
                ret[i].append((len(score) - 2) * score[i][j] - \
                              tdis[i] - tdis[j])
    min_ = 10000000
    index= [-1, -1]
    for i in range(len(score)):
        for j in range(i + 1, len(score)):
            if ret[i][j] < min_:
                min_ = ret[i][j]
                index[0] = i
                index[1] = j
    return index, (tdis[index[0]] - tdis[index[1]]) / (len(score) - 2), \
           score[index[0]][index[1]]

def neighbor(score, inex, num):
    global nu
    if len(score) == 2:
        ret = []
        for i in range(num):
            ret.append([])
        ret[inex[0]].append([inex[1],score[0][1]])
        ret[inex[1]].append([inex[0],score[0][1]])
        return ret
    index, delta, sc = compute_Q(score)
    ini = [inex[index[0]], inex[index[1]]]
    nu_ = nu
    score.append([])
    for i in range(len(score) - 1):
        score[i].append((score[i][index[0]] + score[i][index[1]] - \
                         score[index[0]][index[1]]) / 2)
        score[-1].append(score[i][-1])
    score[-1].append(0)
    score.pop(index[1])
    score.pop(index[0])
    inex.pop(index[1])
    inex.pop(index[0])
    inex.append(nu)
    nu += 1
    for i in range(len(score)):
        score[i].pop(index[1])
        score[i].pop(index[0])
    adj = neighbor(score, inex, num)
    adj[nu_].append([ini[0], (delta + sc) / 2])
    adj[nu_].append([ini[1], (- delta + sc) / 2])
    adj[ini[0]].append([nu_, (delta + sc) / 2])
    adj[ini[1]].append([nu_, ( - delta + sc) / 2])
    return adj

num = int(input())
score = []
for i in range(num):
    score.append(list(map(int, input().split())))
t = []
for i in range(len(score)):
    t.append(i)
nu = len(score)
array = neighbor(score, t, 2 * len(score))
for t in range(len(array)):
    for j in array[t]:
        print(str(t) + "->" + str(j[0]) + ":" + '{:.3f}'.format(round(j[1], 3)))
```
