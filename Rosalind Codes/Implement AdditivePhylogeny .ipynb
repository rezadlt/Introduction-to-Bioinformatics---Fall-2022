```python
def limb(score):
    en = len(score) - 1
    fix = 0
    min_ = 1000000000
    index = 0
    for i in range(1, len(score) - 1):
        x = score[en][i] + score[en][fix] - score[i][fix]
        if x < min_:
            min_ = x
            index = i
    min_ = min_ //2
    return score[en][0] - min_, min_, index

def additive(score, num):
    if len(score) == 2:
        ret = [[[1, score[0][1]]], [[0, score[0][1]]]]
        for i in range(num - 2):
            ret.append([])
        return ret
    dis, min_,index = limb(score)
    sel = len(score) - 1
    score.pop(sel)
    for i in range(len(score)):
        score[i].pop(len(score[i]) - 1)
    adj = additive(score, num)
    visit = [True] * len(adj)
    visit[0] = False
    arr = [0]
    stack = [0]
    while arr[-1] != index:
        val = False
        for i in adj[stack[-1]]:
            if visit[i[0]]:
                val = True
                visit[i[0]] = False
                stack.append(i[0])
        if val:
            arr.append(stack[-1])
        else:
            stack.pop(len(stack) - 1)
            arr.pop(len(arr) - 1)
            if arr[-1] != stack[-1]:
                arr.append(stack[-1])
    arrs = []
    for i in range(len(arr) - 1):
        arrs.append([arr[i], -1])
        for t in adj[arr[i]]:
            if t[0] == arr[i + 1]:
                arrs[i][1] = t[1]
                break
    arrs.append([arr[-1], 0])
    nu = 0
    while True:
        if arrs[nu][1] <= dis:
            dis -= arrs[nu][1]
        else:
            break
        nu += 1
    if dis == 0:
        adj[arrs[nu][0]].append([sel, min_])
        adj[sel].append([arrs[nu][0], min_])
    else:
        adj.append([])
        adj[-1].append([sel, min_])
        adj[sel].append([len(adj) - 1, min_])
        adj[-1].append([arrs[nu][0], dis])
        adj[-1].append([arrs[nu + 1][0], arrs[nu][1] - dis])
        for t in adj[arrs[nu][0]]:
            if t[0] == arrs[nu + 1][0]:
                t[0] = len(adj) - 1
                t[1] = dis
                break
        for t in adj[arrs[nu + 1][0]]:
            if t[0] == arrs[nu][0]:
                t[0] = len(adj) - 1
                t[1] = arrs[nu][1] - dis
                break
    return adj

num = int(input())
score = []
for i in range(num):
    score.append(list(map(int, input().split())))
array = additive(score, len(score))
for t in range(len(array)):
    for j in array[t]:
        print(str(t) + "->" + str(j[0]) + ":" + str(j[1]))
```
